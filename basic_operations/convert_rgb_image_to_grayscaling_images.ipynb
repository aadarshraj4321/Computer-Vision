```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/drive/MyDrive/learning/computer_vision/basic/image/dog.jpg")
```

```python
cv2_imshow(gray_img)
```
