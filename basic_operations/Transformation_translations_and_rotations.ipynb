```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
## Translations

    # This an affine transform that simply shifts the position of an image (left or right)
    # We use cv2.warpAffine to implement these transformations.
    # cv2.warpAffine(image, T, (width, height))

    # T = [1, 0, Tx
    #      0, 1, Ty]

    # Tx : Represents the shift along the x-axis (horizontal)
    # Ty : Represents the shift along the y-axis (vertical)
```

```python
# load the image
img = cv2.imread("/content/drive/MyDrive/learning/computer_vision/basic/image/dog.jpg")
cv2_imshow(img)
```

```python
# store height and width of the image
height, width = img.shape[:2]

# we shift it by quarter of the height and width
quarter_height, quarter_width = height / 4, width / 4

# Our Translation
#      | 1 0 Tx |
# T =  | 0 1 Ty |

# T is our translation matrix
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])

# we use warpAffine to transform the image using the matrix, T
img_translation = cv2.warpAffine(img, T, (width, height))
cv2_imshow(img_translation)
```

```python
## Rotations
## cv2.getRotationMatrix2D(rotation_center_x, rotation_center_y, angle_of_rotation, scale)
```

```python
## Divide by two to rotate the image around its center
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), 90, 1)

# Input our image, the rotation matrix and our desired final width and height
rotated_img = cv2.warpAffine(img, rotation_matrix, (width, height))
cv2_imshow(rotated_img)
```

```python
## Divide by two to rotate the image around its center
## scale is now : 0.5
rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), 90, 0.5)

# Input our image, the rotation matrix and our desired final width and height
rotated_img = cv2.warpAffine(img, rotation_matrix, (width, height))
cv2_imshow(rotated_img)
```
