```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/drive/MyDrive/learning/computer_vision/basic/image/dog.jpg")
```

```python
#cv2_imshow(img)
```

```python
# RGB = Red, Green, Value
# HSV = Hue, Saturation, Value
```

```python
# Use cv2.split to get each color space separately
B, G, R = cv2.split(img)
print(B.shape)
print(G.shape)
print(R.shape)
```

```python
# Each color space on it's on will look like a grascale as it lacks the other color channels
cv2_imshow(B)
```

```python
cv2_imshow(G)
```

```python
cv2_imshow(R)
```

```python
# let's create a matrix of zeros
# with dimensions of the image width * height
zeros = np.zeros(img.shape[:2], dtype = "uint8")

cv2_imshow(cv2.merge([B, zeros, zeros]))
cv2_imshow(cv2.merge([zeros, G, zeros]))
cv2_imshow(cv2.merge([zeros, zeros, R]))
```

```python
# we can merge all of this to form our main image
merged_bgr = cv2.merge([B, G, R])
cv2_imshow(merged_bgr)
```

```python
# HSV = Hue Saturation Value
```

```python
# Hue = 0 - 179
# Saturation = 0 - 255
# Value (Intensity) = 0 - 255
```

```python
## convert BGR to HSV
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv_img)
```

```python
# Let's view each channel type in the HSV Color Space representation
```

```python
# Switching back to viewing the RGB representation
cv2_imshow(hsv_img[:, :, 0])
cv2_imshow(hsv_img[:, :, 1])
cv2_imshow(hsv_img[:, :, 2])
```
