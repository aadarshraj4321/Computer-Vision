```python
## Arithmetic Operations
##      These are simple operations that allows use to directly add or subtract to the color intensity.
##      Calculate the per-element operation of two arrays. The overall effect is increasing or decreasing brightness
```

```python
## Adding comma zero in cv2.imread loads our image in as a grayscaled image
img_gray = cv2.imread("/content/drive/MyDrive/learning/computer_vision/basic/image/dog.jpg",  0)
print(img_gray.shape)
print(img_gray)
#cv2_imshow(img_gray)
print("------------------------------------")

## Create a matrix of ones, then multiply it by a scaler of 100
## THis gives a matrix with same dimensions of out image with all values being 100
M = np.ones(img_gray.shape, dtype = "uint8") * 100
print(M)
```

```python
cv2_imshow(img_gray)
```

```python
## Increasing Brightness
```

```python
## We use this to add this matrix M, to our image
## Notice the increase in brightness
added = cv2.add(img_gray, M)
cv2_imshow(added)

print("-----------------------------------")
## Now if we just added it, look what happens
added2 = img_gray + M
cv2_imshow(added2)
```

```python
## Decreasing Brightness
```

```python
## Likewise we can also subtract
## Notice the decrease in brightness

subtracted = cv2.subtract(img_gray, M)
cv2_imshow(subtracted)

print("---------------------------------")

subtracted2 = img_gray - M
cv2_imshow(subtracted2)
```

```python
## Bitwise operations And Masking
##      To demonstrate these operations let's create some simple images
```

```python
## If you're wondering why only two dimensions, well this a grayscale image,

## Making a square
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)
cv2_imshow(square)

print("--------------------------------")

## Making a ellipse
ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
cv2_imshow(ellipse)
```

```python
## Expermenting with some bitwise operations such as AND, OR, XOR and NOT
```

```python
## Shows only where they intersect
bitwiseAnd = cv2.bitwise_and(square, ellipse)
cv2_imshow(bitwiseAnd)


## Shows where either square or ellipse is
bitwiseOr = cv2.bitwise_or(square, ellipse)
cv2_imshow(bitwiseOr)

## Shows where either exist by itself
bitwiseXor = cv2.bitwise_xor(square, ellipse)
cv2_imshow(bitwiseXor)

## Shows something that isn't part of the square
bitwiseNot_sq = cv2.bitwise_not(square)
cv2_imshow(bitwiseNot_sq)
```
